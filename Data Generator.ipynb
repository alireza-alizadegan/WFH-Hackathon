```python
import numpy as np
import pandas as pd
```

```python
number_of_users = 20
number_of_projects = 10
number_of_interests = 5
```

```python
user_interest_array = []
for user_id in range(number_of_users):
    user_interests_set = set()
    for interest in range(np.random.randint(1, number_of_interests)):
        interest_id = np.random.randint(1, number_of_interests)
        flag = True
        while interest_id in user_interests_set:
                interest_id = np.random.randint(1, number_of_interests)
        user_interests_set.add(interest_id)
        rating = np.random.randint(1, 5)
        experience = np.random.randint(1, 5)
        user_interest_array.append([user_id, interest_id, rating, experience])
        
user_interest_df = pd.DataFrame(data = user_interest_array, columns=['user_id', 'interest_id', 'rating', 'experience'])
user_interest_df
```

```python
user_interest_df.to_csv('user_interest_df.csv', index=True)
```
