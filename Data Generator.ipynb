```python
import numpy as np
import pandas as pd
```

```python
number_of_users = 100
number_of_projects = 50
number_of_roles = 13
number_of_rating = 5
number_of_experience = 3
```

```python
user_role_array = []
for user_id in range(number_of_users + 1):
    user_roles_set = set()
    for role in range(np.random.randint(1, number_of_roles + 1)):
        role_id = np.random.randint(1, number_of_roles + 1)
        flag = True
        while role_id in user_roles_set:
                role_id = np.random.randint(1, number_of_roles + 1)
        user_roles_set.add(role_id)
        rating = np.random.randint(1, number_of_rating + 1)
        experience = np.random.randint(1, number_of_experience + 1)
        user_role_array.append([user_id, role_id, rating, experience])
        
user_role_df = pd.DataFrame(data = user_role_array, columns=['user_id', 'role_id', 'rating', 'experience'])
user_role_df
```

```python
user_role_df.to_csv('user_role_df.csv', index=True)
```

```python
project_role_df.to_csv('project_role_df.csv', index=True)
```
